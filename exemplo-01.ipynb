Neste primeiro exemplo nois iremos calcular a função f(x) = x^2 e sua derivada f'(x) = 2x e usar o matplot para plotar no grafico a curva feita pela função quadratica e a reta feito pela derivada para os valores gerados pelo numpy

Primeiro vamos importas as bibliotecas necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Aqui nos usamos o linspace para gerar um array contendo 1000 valores tendo o seu valor inicial -10 e 10 o valor final

```python
x = np.linspace(-10, 10, 1000)
```

Calculando o array de valores para a funçã f(x) = x^2

```python
y = x**2
# print(y)
```

```python
x_lim = 2
y_lim = x_lim**2 
```

Calculando os valores para a derivada f'(x) = 2x

```python
dy_dx = 2*x
# print(dy_dx)
```

Vamos montar o gráfico para plotar as informações, perceba que vamos plotar 2 linhas tracejadas para indicar ontem acontece a intersecção entre as funções

```python
fig, ax = plt.subplots()

ax.plot(x, y, label='y=x^2')
ax.axvline(x=x_lim, color='red', linestyle='--', label=f'x={x_lim}')
ax.axhline(y=y_lim, color='green', linestyle='--', label=f'y={y_lim}')

ax.plot(x, dy_dx, label='f\'(x)=2x')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Função Quadrática')
ax.legend()
```
